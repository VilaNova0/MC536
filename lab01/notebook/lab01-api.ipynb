```python
import io
import requests
import json
import pandas as pd

url = "https://restcountries.eu/rest/v2/regionalbloc/usan?fields=name;capital;population"
dados = requests.get(url).content
data = json.loads(dados)
tabela = pd.json_normalize(data)
tabela.head(len(tabela))
```

```python
import pandas as pd
import io
import requests
url = "https://www.quandl.com/api/v3/datasets/WIKI/AAPL.csv"
data = requests.get(url).content
csv = pd.read_csv(io.StringIO(data.decode("utf-8")))
print(csv)
```
